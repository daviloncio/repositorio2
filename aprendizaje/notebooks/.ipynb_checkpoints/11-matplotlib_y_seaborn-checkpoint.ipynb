## <a id='coordpar'>Coordenadas paralelas</a>
---------------------------------------

Se suele usar para mostrar diferencias y dispersión entre patrones

```python
# Iris dataset (es una de las bases de datos que están como ejemplo en Seaborn)
df = sns.load_dataset('iris')
df.head(3)
```

```python
plt.figure(figsize=(10,5))
parallel_coordinates(df, 'species', # clase
                     color=["blue", "green", "red"],
#                     colormap=plt.get_cmap("Set2"),
#                     colormap=cm.Paired,
                     alpha=0.5)
plt.legend(fontsize=12)
plt.xticks(fontsize=14);
```

```python
df2 = df.copy()
from sklearn.preprocessing import StandardScaler
aux = df2.values[:,:-1]
scaler = StandardScaler()
#scaler.fit(aux)
#aux_scaled = scaler.transform(aux)
aux_scaled = scaler.fit_transform(aux)
df2.iloc[:,:-1] = aux_scaled
df2.head(5)
```

```python
df2.describe()
```

```python
# ahora normalizando:
plt.figure(figsize=(10,5))
parallel_coordinates(df2, 'species', # clase
                     color=["blue", "green", "red"],
#                     colormap=plt.get_cmap("Set2"),
#                     colormap=cm.Paired,
                     alpha=0.5)
plt.legend(fontsize=12)
plt.xticks(fontsize=14);
```

```python
# Wine dataset
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
data = load_wine()
data.data
```

```python
sc = StandardScaler()
df = pd.DataFrame(sc.fit_transform(data.data), columns=data.feature_names)
df["class"] = data.target_names[data.target]
plt.figure(figsize=(14,5))
parallel_coordinates(df, 'class', color=["blue", "green", "red"],
                     alpha=0.5) #colormap=cm.Paired)
plt.xticks(rotation=85, fontsize=12)
plt.show()
df[:3]
```

## <a id='radarplot'>Radar plot</a>
---------------------------------------

Se suele usar para mostrar diferencias entre unos pocos patrones individuales, o para mostrar diferencias entre las medias de las clases.

### Radar plot en base de datos Iris

```python
#!pip install radar_chart
```

```python
from radar_chart import radar_factory

df = sns.load_dataset('iris')
df.head(100)
```

```python
aux = df.groupby("species").mean()
aux
```

```python
nombres_atributos = aux.columns
n_atributos = len(nombres_atributos)
theta = radar_chart(n_atributos, frame='circle')
```

```python
nombres_atributos
```

**Promedio de atributos en cada clase:**

```python
fig, axes = plt.subplots(figsize=(5, 5), nrows=1, ncols=1,
                         subplot_kw=dict(projection='radar'))
fig.subplots_adjust(wspace=0.25, hspace=0.20, top=0.85, bottom=0.05)

colors = ['b', 'r', 'g', 'm', 'y']
# Plot the three cases from the example data on separate axes
ax=axes
#ax.set_rgrids([0.2, 0.4, 0.6, 0.8])
ax.set_title("Medias intraclase", weight='bold', size='medium', position=(0.5, 1.1),
             horizontalalignment='center', verticalalignment='center')

clases = aux.index # nombres de las filas de aux
for clase, color in zip(clases, colors):
    ax.plot(theta, aux.loc[clase], color=color)
#    ax.fill(theta, aux.loc[clase], facecolor=color, alpha=0.25)
    ax.set_varlabels(nombres_atributos)

# add legend relative to top-left plot
ax.legend(nombres_atributos, loc=(0.9, .95),
          labelspacing=0.1, fontsize='small');
```

### Radar plot en base de datos de vinos

```python
data = load_wine()
sc = StandardScaler()
df = pd.DataFrame(sc.fit_transform(data.data), columns=data.feature_names)
df["class"] = data.target_names[data.target]
df.head(3)
```

```python
aux = df.groupby("class").mean()
aux
```

```python
nombres_atributos = aux.columns
n_atributos = len(nombres_atributos)
theta = radar_factory(n_atributos, frame='circle')
```

**Promedio de atributos en cada clase:**

```python
fig, axes = plt.subplots(figsize=(5, 5), nrows=1, ncols=1,
                         subplot_kw=dict(projection='radar'))
fig.subplots_adjust(wspace=0.25, hspace=0.20, top=0.85, bottom=0.05)

colors = ['b', 'r', 'g', 'm', 'y']
# Plot the four cases from the example data on separate axes
ax=axes
#ax.set_rgrids([0.2, 0.4, 0.6, 0.8])
ax.set_title("Medias intraclase", weight='bold', size='medium', position=(0.5, 1.1),
             horizontalalignment='center', verticalalignment='center')

clases = aux.index # nombres de las filas de aux
for clase, color in zip(clases, colors):
    ax.plot(theta, aux.loc[clase], color=color)
    ax.fill(theta, aux.loc[clase], facecolor=color, alpha=0.5)
    ax.set_varlabels(nombres_atributos)

# add legend relative to top-left plot
labels = aux.index
ax.legend(labels, loc=(0.9, .95),
          labelspacing=0.1, fontsize='small');
```

## <a id='diagbur'>Diagrama de burbujas</a>
---------------------------------------

```python
# create data
x = np.random.rand(15)
y = x+np.random.rand(15)
radio = 2000*np.random.rand(15)
color = np.random.rand(15)
 
# Change color with c and alpha. I map the color to the X axis value.
plt.scatter(x, y, s=radio, c=color,
            cmap="cool", alpha=0.4, edgecolors="grey", linewidth=2)
 
# Add titles (main and on axis)
plt.xlabel("the X axis")
plt.ylabel("the Y axis")
plt.title("A colored bubble plot");
```

## <a id='stacked_areas'>Áreas apiladas</a>
---------------------------------------

```python
# Dataset
baseline = ["zero", "sym", "wiggle", "weighted_wiggle"]
np.random.seed(1234567)
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])

#df.plot.area()
plt.figure(figsize=(10,10))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.stackplot(range(10), df.values.T,
                  baseline=baseline[i], labels=df.columns, alpha=0.9)
    plt.title(baseline[i])
    plt.legend(loc=2)
plt.show()
df
```

## <a id='stacked_areas_perc'>Áreas apiladas (porcentaje)</a>
---------------------------------------

```python
# Make data
data = pd.DataFrame({'group_A':[1,4,6,8,9],
                     'group_B':[2,24,7,10,12],
                     'group_C':[2,8,5,10,6], },
                    index=range(1,6))
data
```

```python
suma_por_fila = data.sum(axis=1)
suma_por_fila
```

```python
data.divide(suma_por_fila, axis=0)
```

```python
# We need to transform the data from raw data to percentage (fraction)
data_perc = data.divide(suma_por_fila, axis=0)
 
# Make the plot
x = range(1,6)
plt.stackplot(x,  data_perc["group_A"],  data_perc["group_B"], 
              data_perc["group_C"], labels=['A','B','C'])
plt.legend(loc='upper left')
plt.margins(0,0)
plt.xticks(x)
plt.title('100 % stacked area chart');
```

## <a id='diagramas_donut'>Diagramas de donut</a>
---------------------------------------

**Recomendado en vez del diagrama de tarta: los diagramas de donut resaltan la longitud del arco de cada sector, y ahí sí somos buenos distinguiendo proporciones entre longitudes de arcos.**

```python
# create data
names='groupA', 'groupB', 'groupC', 'groupD',
size=[12,11,3,30]
 
# figure:
plt.figure(figsize=(5,5))
plt.pie(size, labels=names, colors=['red','green','blue','skyblue'],
        wedgeprops=dict(width=0.3)
       );
```

```python
# pip install palettable
# https://jiffyclub.github.io/palettable/

from palettable.colorbrewer.qualitative import Pastel1_7

plt.figure(figsize=(5,5))
plt.pie(size, labels=names, colors=Pastel1_7.hex_colors,
        wedgeprops=dict(width=0.3)
       );
```

```python
# Label distance: gives the space between labels and the center of the pie
plt.figure(figsize=(5,5))
plt.pie(size, labels=names, labeldistance=0.45,
        wedgeprops=dict(width=0.3)
       )
plt.rcParams['text.color'] = 'red';
```

```python
# Custom wedges
plt.figure(figsize=(6,6))
plt.rcParams['text.color'] = 'black'
plt.pie(size, labels=names,
        wedgeprops = dict(width=0.3, linewidth=7, edgecolor='white'))
p=plt.gcf()
my_circle=plt.Circle( (0,0), 0.7, color='white')
p.gca().add_artist(my_circle)
plt.show()
```

```python
# Custom wedges
f = plt.figure(figsize=(6,6))
f.patch.set_facecolor('black')
plt.rcParams['text.color'] = 'white'
plt.pie(size, labels=names,
        wedgeprops=dict(width=0.3)
       );
```

```python
# Make data: I have 3 groups and 7 subgroups
group_names=['groupA', 'groupB', 'groupC']
group_size=[12,11,30]
subgroup_names=['A.1', 'A.2', 'A.3', 'B.1', 'B.2', 'C.1', 'C.2', 'C.3', 'C.4', 'C.5']
subgroup_size=[4,3,5,6,5,10,5,5,4,6]
 
# Create colors
a, b, c=[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]
 
# First Ring (outside)
fig, ax = plt.subplots()
plt.rcParams['text.color'] = 'black'
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=1.3, labels=group_names, colors=[a(0.6), b(0.6), c(0.6)] )
plt.setp( mypie, width=0.3, edgecolor='white')
 
# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=1.3-0.3, labels=subgroup_names,
                   labeldistance=0.7,
                   colors=[a(0.5), a(0.4), a(0.3),
                           b(0.5), b(0.4),
                           c(0.6), c(0.5), c(0.4), c(0.3), c(0.2)])
plt.setp(mypie2, width=0.4, edgecolor='white');
#plt.margins(0,0);
```

## <a id='diagramas_violin'>Diagramas de violín</a>
---------------------------------------

### Diagramas de violín en la base de datos Iris

```python
df = sns.load_dataset('iris')
plt.figure(figsize=(5,5))
sns.violinplot(x='species', y='sepal_length', data=df,
               order=[ "versicolor", "virginica", "setosa"]);
```

**Ordenado por medianas:**

```python
df.groupby(by=["species"])["sepal_length"].median().sort_values()
```

```python
orden = df.groupby(by=["species"])["sepal_length"].median().sort_values().index
orden
```

```python
plt.figure(figsize=(5,5))
sns.violinplot(x='species', y='sepal_length', data=df,
               order=orden);
```

## <a id='diagramas_venn'>Diagramas de Venn</a>
---------------------------------------

```python
# pip install matplotlib-venn

from matplotlib_venn import venn2, venn3

set1 = set(["A", "B", "C", "D"])
set2 = set(["A", "G", "H"])
set3 = set(["A", "Z"])

plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
v = venn2([set1, set2], alpha = 1, set_labels=("set 1", "set 2"))
plt.subplot(1,3,2)
v = venn3([set1, set2, set3], alpha = 1, set_labels=("set 1", "set 2", "set 3"))
plt.subplot(1,3,3)
v = venn3([3,2,0,1,0,0,1], alpha = 1, set_labels=("set 1", "set 2", "set 3"))
plt.show()
```
