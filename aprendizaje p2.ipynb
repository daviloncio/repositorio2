Ahora procedemos a eliminar aquellas variables colineales, las detectaremos al ver en la matriz de correlaci贸n valores superiores a | 0.8 |, ya que nos ha parecido un valor 贸ptimo para ver las colineales. Tiene que ser valor absoluto ya que pueder ser colineales y tener correlaci贸n negativa.

```python
matriz_cor=data.corr()
data.info()
matriz_cor
```

```python
#si detectamos una correlaci贸n entre dos variables, tenemos que borrar ambas?
print(type(data))
numsmasque08=[]
list_columnas_a_eliminar=[]
indice_col=0 ; indice_fila=0
import numpy as np ;
for tupla in matriz_cor.iterrows(): #saco variable
    indice_col += 1
    indice_fila=0  #iteramos sobre una nueva columna
    serie=tupla[1]
    nombre_columna=tupla[0]   
    for num in serie:
        indice_fila += 1
        if abs(num)>=0.8:
            if num==1 and indice_fila==indice_col:  #no nos interesa quitar las columnas con 1.00 si este esta en la diagonal
                pass
            else:
                numsmasque08.append(num)
                list_columnas_a_eliminar.append(nombre_columna)
                
                   #eliminamos la columna/fila de matriz_cor o de data?
                
for nombre in list_columnas_a_eliminar:
    del(data[nombre])



   #hay 15 nums cuyo valor absoluto es mayor que 0.8

```
